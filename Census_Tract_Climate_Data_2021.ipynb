### Create a function to deal with missing data

```python
def changeToNone(value):
    if value == 999.9 or value == 9999.9 or value == 99.99 or value == 99.9:
        return np.nan
    else:
        return value
```

```python
def changeMissingData(df):
    df['TEMP'] = df['TEMP'].apply(changeToNone)   # missing data = 9999.9
    df['DEWP'] = df['DEWP'].apply(changeToNone)   # missing data = 9999.9
    df['SLP'] = df['SLP'].apply(changeToNone)     # missing data = 9999.9
    df['STP'] = df['STP'].apply(changeToNone)     # missing data = 9999.9
    df['STP'] = df['STP'].apply(changeToNone)     # missing data = 999.9
    df['VISIB'] = df['VISIB'].apply(changeToNone) # missing data = 999.9
    df['WDSP'] = df['WDSP'].apply(changeToNone)   # missing data = 999.9
    df['MXSPD'] = df['MXSPD'].apply(changeToNone) # missing data = 999.9
    df['GUST'] = df['GUST'].apply(changeToNone)   # missing data = 999.9
    df['MAX'] = df['MAX'].apply(changeToNone)     # missing data = 9999.9
    df['MIN'] = df['MIN'].apply(changeToNone)     # missing data = 9999.9
    df['PRCP'] = df['PRCP'].apply(changeToNone)   # missing data = 99.99
    df['SNDP'] = df['SNDP'].apply(changeToNone)   # missing data = 999.9
    return df
```

## Calculate Relative Humidity

```python
def Convertion_F_to_C(T):
    return (T - 32)*(5/9)
```

```python
import math
```

```python
def RH_Calculator(T, TD):
    T = Convertion_F_to_C(T)
    TD = Convertion_F_to_C(TD)
    return 100*(math.exp((17.625*TD)/(243.04+TD))/math.exp((17.625*T)/(243.04+T)))
```

```python
df['Relative_Humidity'] = df.apply(lambda x: RH_Calculator(x['mean_temp'], x['dewpoint']), axis=1)
```

### Calculate Heat Index

```python
def heat_index_calculation(T, RH):
    if RH < 13 and T > 80 and T < 112:
        return ((13-RH)/4)* math.sqrt((17- abs(T-95))/17)
    elif RH > 85 and T > 80 and T < 87:
        return ((RH - 85)/10)*((87-T)/5)
    elif T < 80:
        return 0.5 * (T + 61.0 + ((T-68.0)*1.2) + (RH*0.094))
    else:
        return -42.379 + 2.04901523*T + 10.14333127*RH - .22475541*T*RH - .00683783*T*T - .05481717*RH*RH + .00122874*T*T*RH + .00085282*T*RH*RH - .00000199*T*T*RH*RH
```

```python
df['Heat_Index'] = df.apply(lambda x: heat_index_calculation(x['mean_temp'], x['Relative_Humidity']), axis=1)
```

```python
def generate_soup(url):
    response = requests.get(url)
    html = response.text
    return BeautifulSoup(html, "html.parser")
```

```python
def findFileList(url, year):
    newUrl = url + year + '/'
    soup = generate_soup(newUrl)
    files = soup.find_all('a')
    files = files[5:]
    fileList = []
    for ele in files:
        form = ele.text
        fileList.append(form)
    return fileList
```

```python
def createUrls(fileList, year):
    urls = []
    for element in fileList:
        if (element in currentList):
            url = "https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/" + year + "/" + element
            urls.append(url)
    return urls
```

```python
def findDataFrame(url):
    r = requests.get(url)
    open('temp.csv', 'wb').write(r.content)
    df = pd.read_csv('temp.csv', dtype = {'FRSHTT': str})
    if df.empty != True:
        df = changeMissingData(df)
        df = renameColumns(df)
        df['FRSHTT'] = df['FRSHTT'].apply(changeDataType)
        df = seperateFRSHTT(df)
        df['Relative_Humidity'] = df.apply(lambda x: RH_Calculator(x['mean_temp'], x['dewpoint']), axis=1)
        df['Heat_Index'] = df.apply(lambda x: heat_index_calculation(x['mean_temp'], x['Relative_Humidity']), axis=1)
        df['date'] = pd.to_datetime(df['date'])
        df['year'] = df['date'].dt.year
        df['Month'] = df['date'].apply(getMonth)
        df = selectColumns(df)
    return df
```

```python
def groupByMonth(df):
    df_part1 = df[['station_id', 'date', 'station_lat', 'station_lon', 'elevation','station_name', 'mean_temp', 'dewpoint', 'sea_level_pressure','station_pressure', 'visibility', 'wind_speed', 'precipitation', 'Relative_Humidity','Heat_Index', 'year', 'Month']]
    df_part2 = df[['station_id', 'date', 'station_lat', 'station_lon', 'elevation','station_name', 'fog','rain', 'snow', 'hail', 'thunder', 'tornado', 'year', 'Month']]
    df_part1 = df_part1.groupby(['station_id', 'station_lat', 'station_lon', 'station_name', 'elevation','Month']).mean()
    df_part2 = df_part2.groupby(['station_id', 'station_lat', 'station_lon', 'station_name', 'elevation','Month']).sum()
    df_part1 = df_part1.reset_index()
    df_part2 = df_part2.reset_index()
    df = df_part1.merge(df_part2, on = ['station_id', 'station_lat', 'station_lon', 'elevation','station_name', 'Month'], how = 'left')
    df = df.drop(['year_y'], axis = 1)
    df = df.rename(columns = {'year_x': 'year'})
    df['year'] = df['year'].astype(str)
    return df
```

```python
def groupByYear(df):
    df_part1 = df[['station_id', 'date', 'station_lat', 'station_lon', 'elevation','station_name', 'mean_temp', 'dewpoint', 'sea_level_pressure','station_pressure', 'visibility', 'wind_speed', 'precipitation', 'Relative_Humidity','Heat_Index', 'year', 'Month']]
    df_part2 = df[['station_id', 'date', 'station_lat', 'station_lon', 'elevation','station_name', 'fog','rain', 'snow', 'hail', 'thunder', 'tornado', 'year', 'Month']]
    df_part1 = df_part1.groupby(['station_id', 'station_lat', 'station_lon', 'station_name', 'elevation','year']).mean()
    df_part2 = df_part2.groupby(['station_id', 'station_lat', 'station_lon', 'station_name', 'elevation','year']).sum()
    df_part1 = df_part1.reset_index()
    df_part2 = df_part2.reset_index()
    df = df_part1.merge(df_part2, on = ['station_id', 'station_lat', 'station_lon', 'elevation','station_name', 'year'], how = 'left')
    return df
```

```python
def combineYearData(urls):
    df1 = pd.DataFrame()
    for url in urls:
        df2 = findDataFrame(url)
        df2 = groupByYear(df2)
        df = pd.concat([df1, df2], axis = 0)
        df1 = df
    return df1
```

```python
def dist(p0, p1):
    return (((p0[0] - p1[0])**2) + ((p0[1] - p1[1])**2))**.5
```

```python
def findSquare(lat, lon):
    square = []
    top = lat + 0.5
    down = lat - 0.5
    right = lon + 0.5
    left = lon - 0.5
    square.append(top)
    square.append(down)
    square.append(left)
    square.append(right)
    return square
```

```python
def findPoints(lat, lon):
    square = findSquare(lat, lon)
    top = square[0]
    down = square[1]
    left = square[2]
    right = square[3]
    points = []
    for point in pointList:
        p_lat = point[1]
        p_lon = point[2]
        if point[1] <= top and point[1] >= down and point[2] >= left and point[2] <= right:
            points.append(point)
            
    return points
```

```python
def shortestDistance(lat, lon):
    points = findPoints(lat, lon)
            
    distants = []
    p1 = []
    p1.append(lat)
    p1.append(lon)
    for point in points:
        p_id = point[0]
        p_lat = point[1]
        p_lon = point[2]
        p2 = []
        p2.append(p_lat)
        p2.append(p_lon)
        distant = dist(p1, p2)
        dist_id = []
        dist_id.append(p_id)
        dist_id.append(distant)
        distants.append(dist_id)
    distantList = []    
    for ele in distants:
        distantList.append(ele[1])
    for item in distants:
        if item[1] == min(distantList):
            return item[0]
```
